```python
apps_df=pd.read_csv('Play Store Data_3.csv')
```

```python
apps_df.head()
```

```python
#pd.read_csv() : csv files
#pd.read_excel() : excel files
#pd.read_sql() : SQL Databases
#pd.read_json() : JSON Files
```

```python
#df.isnull() : Missing values
#df.dropna() : Removes rows and columns that contain the missing values
#df.fillna() : Fills missing values
```

```python
#Step 2 : Data Cleaning
apps_df = apps_df.dropna(subset=['Rating'])
for column in apps_df.columns :
    apps_df[column].fillna(apps_df[column].mode()[0],inplace=True)
apps_df.drop_duplicates(inplace=True)
apps_df=apps_df=apps_df[apps_df['Rating']<=5]
```

```python
apps_df.dtypes
```

```python
#Convert the Installs columns to numeric by removing commas and +
apps_df['Installs']=apps_df['Installs'].str.replace(',','').str.replace('+','').astype(int)
```

```python
#Convert Price column to numeric after removing $
apps_df['Price']=apps_df['Price'].str.replace('$','').astype(float)
```

```python
apps_df.dtypes
```

```python
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'k' in size:
        return float(size.replace('k',''))/1024
    else:
        return np.nan
apps_df['Size']=apps_df['Size'].apply(convert_size)
```

```python
apps_df
```

```python
#Lograrithmic
apps_df['Log_Installs']=np.log(apps_df['Installs'])
```

```python
apps_df['Reviews']=apps_df['Reviews'].astype(int)
```

```python
apps_df['Log_Reviews']=np.log(apps_df['Reviews'])
```

```python
apps_df.dtypes
```

```python
apps_df = apps_df[apps_df['Installs'] > 10000]
```

```python
apps_df
```

```python
apps_df = apps_df[apps_df['Android Ver'].str.contains('4.0|4.1|4.2|4.3|4.4|5.0|6.0|7.0|8.0|9.0')]
```

```python
apps_df.head()
```

```python
# Removing apps with size less than 15M.

apps_df = apps_df[apps_df['Size'] > 15]
```

```python
apps_df.head()
```

```python
# Filtering Content Rating to 'Everyone'
apps_df = apps_df[apps_df['Content Rating'] == 'Everyone']
```

```python
apps_df.head()
```

```python
# Filtering out apps with names longer than 30 characters
apps_df = apps_df[apps_df['App'].str.len() <= 30]
```

```python
apps_df.head()
```

```python
# Step 2: Adding Revenue column (Installs * Price)
apps_df['Revenue'] =apps_df['Installs'] * apps_df['Price']
```

```python
apps_df.head()
```

```python
# Filtering apps with revenue above $10,000
apps_df = apps_df[apps_df['Revenue'] > 10000]
```

```python
apps_df.head()
```

```python
# Step 3: Categorizing apps as Free or Paid based on the Type column
apps_df['App Type'] = apps_df['Type'].apply(lambda x: 'Free' if x == 'Free' else 'Paid')
```

```python
# Step 4: Finding top 3 app categories by average installs
top_categories = apps_df.groupby('Category')['Installs'].mean().nlargest(3).index
```

```python
apps_df.head()
```

```python
# Filtering the data for the top 3 categories
df_top_categories = apps_df[apps_df['Category'].isin(top_categories)]
```

```python
apps_df
```

```python
# Aggregating data to get average installs and revenue by app type (Free vs Paid) within the top categories
df_agg = df_top_categories.groupby(['Category', 'App Type']).agg({
    'Installs': 'mean',
    'Revenue': 'mean'
    }).reset_index()
```

```python
apps_df
```

```python
# Step 5: Creating a dual-axis chart
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Creating a figure with two y-axes
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Adding traces for installs (left y-axis)
for app_type in ['Free', 'Paid']:
    filtered_data = df_agg[df_agg['App Type'] == app_type]
    fig.add_trace(
        go.Bar(x=filtered_data['Category'], y=filtered_data['Installs'], name=f'Avg Installs ({app_type})'),
        secondary_y=False,
    )
# Add traces for revenue (right y-axis)
for app_type in ['Free', 'Paid']:
    filtered_data = df_agg[df_agg['App Type'] == app_type]
    fig.add_trace(
        go.Scatter(x=filtered_data['Category'], y=filtered_data['Revenue'], mode='lines+markers', name=f'Avg Revenue ({app_type})'),
        secondary_y=True,
    )
html_files_path="./"
if not os.path.exists(html_files_path):
    os.makedirs(html_files_path)

# Save each Plotly figure to an HTML file
def save_plot_as_html(fig, filename, insight):
        global plot_containers
        filepath = os.path.join(html_files_path, filename)
        html_content = pio.to_html(fig, full_html=False, include_plotlyjs='inline')
        # Append the plot and its insight to plot_containers
        plot_containers += f"""
        <div class="plot-container" id="{filename}" onclick="openPlot('{filename}')">
            <div class="plot">{html_content}</div>
            <div class="insights">{insight}</div>
        </div>
        """
        fig.write_html(filepath, full_html=False, include_plotlyjs='inline')
```

```python
# Updating layout
fig.update_layout(
    title='Comparison of Average Installs and Revenue for Free vs Paid Apps in Top 3 Categories',
    xaxis_title='App Category',
    yaxis_title='Average Installs',
    yaxis2_title='Average Revenue ($)',
    legend_title='App Type',
    barmode='group'
)


plot_containers=""
```
